```python
# Panda2.quest
df = pd.read_csv('states.areacodes.csv')
df
```

```python
# remove the 'Washington,DC' which is not a state
df = df[df['State']!='Washington,DC']
df
```

```python
# 用groupby统计次数，as_index = False 代表还是用回原来的index
df_group = df.groupby(['State'], sort = True , as_index=False)
df_count = df_group.count()
```

```python
# drop a column
newdf = df_count.drop(columns='Area code')
newdf
```

```python
# rename the column
newdf2 = newdf.rename(columns={'State code':'times'}, index=str)
newdf2
```

```python
newdf3 = newdf2[newdf2['times']==1]
newdf3
```

```python
# Part B
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('zipcodes.states.gps.csv')
df
```

```python
df2 = df.groupby(['state'], sort=True, as_index=False)
df3 = df2.count()
df3
```

```python
df4 = df3.drop(columns=['latitude','longitude','county','city'])
df5 = df4.rename(index=str, columns={'zip_code':'times'})
df6 = df5.sort_values(ascending=False, by=['times'])
```

```python
y1 = list(df6['times'])
x1 = list(df6['state'])
# plt.plot(y=y1, x=x1, kind = 'bar')
x_pos = [i for i, _ in enumerate(x1)]

plt.bar(x_pos,y1,color = 'red', )
plt.xticks(x_pos, x1)
plt.xticks(fontsize=5)
plt.figure(figsize=(2,1))
```
